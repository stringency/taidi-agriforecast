```python
#传入农产品名称映射值和其对应的市场映射值，返回一个大量训练数据的列表（格式为DataFrame）
#将935种数据集分类，以便于后面进行每种的插值和训练
train_data = data_farming2_df_screenF1[['农产品市场所在省份','农产品类别','农产品名称映射值','市场名称映射值','数据发布时间','最低交易价格','平均交易价格','最高交易价格']].groupby(by=['农产品市场所在省份','农产品类别','农产品名称映射值','市场名称映射值','数据发布时间'])
train_group_tmp = train_data.agg({"最低交易价格":np.sum, "平均交易价格":np.sum, "最高交易价格":np.sum})#每个数据只有一个，求和仍是自己

train_group_tmp.to_csv('tmp/train_group_tmp.csv',encoding='gbk')#保存分类数据为csv
train_group_tmp2 = pd.read_csv('tmp/train_group_tmp.csv',encoding='gbk')
train_group_tmp2_df = pd.DataFrame(train_group_tmp2)#在此处选出需要插值训练的每种菜品


#异常值用空值代替方法
def replace_abnormal_low(x):
    import numpy as np
    global train_group_tmp_one
    QU = x.quantile(0.75)
    QL = x.quantile(0.25)
    IQR = QU - QL
#     x.loc[(x > (QU + 1.5*IQR)) | (x < (QL-1.5*IQR))] = np.nan
    for i in range(0,len(x)):
        if((x[i]>QU + 1.5*IQR)|(x[i]<QL-1.5*IQR)):
            train_group_tmp_one.loc[i,('最低交易价格')] = np.nan
def replace_abnormal_ave(y):
    import numpy as np
    global train_group_tmp_one
    QU = y.quantile(0.75)
    QL = y.quantile(0.25)
    IQR = QU - QL
#     y.loc[(y > (QU + 1.5*IQR)) | (y < (QL-1.5*IQR))] = np.nan
    for i in range(0,len(y)):
        if((y[i]>QU + 1.5*IQR)|(y[i]<QL-1.5*IQR)):
            train_group_tmp_one.loc[i,('平均交易价格')] = np.nan
def replace_abnormal_high(z):
    import numpy as np
    global train_group_tmp_one
    QU = z.quantile(0.75)
    QL = z.quantile(0.25)
    IQR = QU - QL
#     z.loc[(z > (QU + 1.5*IQR)) | (z < (QL-1.5*IQR))] = np.nan
    for i in range(0,len(z)):
        if((z[i]>QU + 1.5*IQR)|(z[i]<QL-1.5*IQR)):
            train_group_tmp_one.loc[i,('最高交易价格')] = np.nan

def Spline_Interpolation(agrprdValue,markValue):
    import numpy as np
    import matplotlib.pyplot as plt
    from scipy import interpolate
    #获取需要插值训练的数据
    global train_group_tmp_one
    train_group_tmp_one = pd.DataFrame(columns=train_group_tmp2.columns)
    for i in train_group_tmp2_df.itertuples(index=False):#筛选出一种需要训练插值的菜品
        if (getattr(i, "农产品名称映射值")==agrprdValue)&(getattr(i, "市场名称映射值")==markValue):
            train_group_tmp_one.loc[len(train_group_tmp_one.index)] = i
    train_group_tmp_one.sort_values(by = '数据发布时间',inplace=True)#对数据进行按照时间顺序排序
    train_group_tmp_one['数据发布时间'] = pd.to_datetime(train_group_tmp_one['数据发布时间'])#时间格式转化
    #最低交易价格数据异常检测
    replace_abnormal_low(train_group_tmp_one['最低交易价格'])
    #平均交易价格数据异常检测
    replace_abnormal_ave(train_group_tmp_one['平均交易价格'])
    # #最高交易价格数据异常检测
    replace_abnormal_high(train_group_tmp_one['最高交易价格'])
    train_group_tmp_one.dropna(inplace=True)#删去存在异常值的数据
    train_group_tmp_one.reset_index(drop=True,inplace=True)#调整索引
    #自变量初始化处理
    x_time_df = train_group_tmp_one['数据发布时间']
    tgto = train_group_tmp_one['数据发布时间'] - train_group_tmp_one['数据发布时间'][len(train_group_tmp_one.index)-1] + (train_group_tmp_one['数据发布时间'][len(train_group_tmp_one.index)-1]- train_group_tmp_one['数据发布时间'][0])#计算出timedelta 
    x_time = tgto.astype('timedelta64[D]').astype(float)#将dimedelta转变为float，这个为因变量
    
    #因变量初始化处理
    y_low_price,y_ave_price,y_high_price = train_group_tmp_one['最低交易价格'],train_group_tmp_one['平均交易价格'],train_group_tmp_one['最高交易价格']#获取三种因变量值
    
    #自变量补充并且转化格式
    xnew_time = np.arange(train_group_tmp_one['数据发布时间'][0],train_group_tmp_one['数据发布时间'][len(train_group_tmp_one.index)-1]+pd.Timedelta(days=1),pd.Timedelta(days=1))#插入缺失值的自变量
    #由此化解m8[ns]无法和float运算的问题
    xnew_time_df = pd.DataFrame({'数据发布时间':xnew_time})#np.datetime64变为数据框结构
    xnew_time_df['数据发布时间'] = pd.to_datetime(xnew_time_df['数据发布时间'])#转变为pd.datetime
    xtd = xnew_time_df['数据发布时间'] - xnew_time_df['数据发布时间'][len(xnew_time_df.index)-1] + (train_group_tmp_one['数据发布时间'][len(train_group_tmp_one.index)-1]- train_group_tmp_one['数据发布时间'][0])#计算出timedelta
    xnew_timeF = xtd.astype('timedelta64[D]').astype(float)#timedelta转变为float

    #最低交易价格插值
    f_low_price = interpolate.interp1d(x_time, y_low_price)#最低价格的函数
    ynew_low_price = f_low_price(xnew_timeF)#利用interold进行最低价格的插值

    #平均交易价格插值
    f_ave_price = interpolate.interp1d(x_time, y_ave_price)#最低价格的函数
    ynew_ave_price = f_ave_price(xnew_timeF)#利用interold进行最低价格的插值

    #最高交易价格插值
    f_high_price = interpolate.interp1d(x_time, y_high_price)#最低价格的函数
    ynew_high_price = f_high_price(xnew_timeF)#利用interold进行最低价格的插值

    train_data_final = xnew_time_df
    train_data_final.insert(len(xnew_time_df.columns),value=ynew_low_price,column = '最低交易价格')
    train_data_final.insert(len(xnew_time_df.columns),value=ynew_ave_price,column = '平均交易价格')
    train_data_final.insert(len(xnew_time_df.columns),value=ynew_high_price,column = '最高交易价格')
    return train_data_final#返回一个已经插值的需要训练的一种数据
```

```python
def macLear_Neural_Network(agrprdValue,markValue):
    #传入农产品名称映射值和其对应的市场映射值，返回一个大量训练数据的列表（格式为DataFrame）
    #将935种数据集分类，以便于后面进行每种的插值和训练
    train_data = data_farming2_df_screenF1[['农产品市场所在省份','农产品类别','农产品名称映射值','市场名称映射值','数据发布时间','最低交易价格','平均交易价格','最高交易价格']].groupby(by=['农产品市场所在省份','农产品类别','农产品名称映射值','市场名称映射值','数据发布时间'])
    train_group_tmp = train_data.agg({"最低交易价格":np.sum, "平均交易价格":np.sum, "最高交易价格":np.sum})#每个数据只有一个，求和仍是自己

    train_group_tmp.to_csv('tmp/train_group_tmp.csv',encoding='gbk')#保存分类数据为csv
    train_group_tmp2 = pd.read_csv('tmp/train_group_tmp.csv',encoding='gbk')
    train_group_tmp2_df = pd.DataFrame(train_group_tmp2)#在此处选出需要插值训练的每种菜品


    #异常值用空值代替方法
    def replace_abnormal_low(x):
        import numpy as np
        global train_group_tmp_one
        QU = x.quantile(0.75)
        QL = x.quantile(0.25)
        IQR = QU - QL
    #     x.loc[(x > (QU + 1.5*IQR)) | (x < (QL-1.5*IQR))] = np.nan
        for i in range(0,len(x)):
            if((x[i]>QU + 1.5*IQR)|(x[i]<QL-1.5*IQR)):
                train_group_tmp_one.loc[i,('最低交易价格')] = np.nan
    def replace_abnormal_ave(y):
        import numpy as np
        global train_group_tmp_one
        QU = y.quantile(0.75)
        QL = y.quantile(0.25)
        IQR = QU - QL
    #     y.loc[(y > (QU + 1.5*IQR)) | (y < (QL-1.5*IQR))] = np.nan
        for i in range(0,len(y)):
            if((y[i]>QU + 1.5*IQR)|(y[i]<QL-1.5*IQR)):
                train_group_tmp_one.loc[i,('平均交易价格')] = np.nan
    def replace_abnormal_high(z):
        import numpy as np
        global train_group_tmp_one
        QU = z.quantile(0.75)
        QL = z.quantile(0.25)
        IQR = QU - QL
    #     z.loc[(z > (QU + 1.5*IQR)) | (z < (QL-1.5*IQR))] = np.nan
        for i in range(0,len(z)):
            if((z[i]>QU + 1.5*IQR)|(z[i]<QL-1.5*IQR)):
                train_group_tmp_one.loc[i,('最高交易价格')] = np.nan

    def Spline_Interpolation(agrprdValue,markValue):
        import numpy as np
        import matplotlib.pyplot as plt
        from scipy import interpolate
        #获取需要插值训练的数据
        global train_group_tmp_one
        train_group_tmp_one = pd.DataFrame(columns=train_group_tmp2.columns)
        for i in train_group_tmp2_df.itertuples(index=False):#筛选出一种需要训练插值的菜品
            if (getattr(i, "农产品名称映射值")==agrprdValue)&(getattr(i, "市场名称映射值")==markValue):
                train_group_tmp_one.loc[len(train_group_tmp_one.index)] = i
        train_group_tmp_one.sort_values(by = '数据发布时间',inplace=True)#对数据进行按照时间顺序排序
        train_group_tmp_one['数据发布时间'] = pd.to_datetime(train_group_tmp_one['数据发布时间'])#时间格式转化
        #最低交易价格数据异常检测
        replace_abnormal_low(train_group_tmp_one['最低交易价格'])
        #平均交易价格数据异常检测
        replace_abnormal_ave(train_group_tmp_one['平均交易价格'])
        # #最高交易价格数据异常检测
        replace_abnormal_high(train_group_tmp_one['最高交易价格'])
        train_group_tmp_one.dropna(inplace=True)#删去存在异常值的数据
        train_group_tmp_one.reset_index(drop=True,inplace=True)#调整索引
        #自变量初始化处理
        x_time_df = train_group_tmp_one['数据发布时间']
        tgto = train_group_tmp_one['数据发布时间'] - train_group_tmp_one['数据发布时间'][len(train_group_tmp_one.index)-1] + (train_group_tmp_one['数据发布时间'][len(train_group_tmp_one.index)-1]- train_group_tmp_one['数据发布时间'][0])#计算出timedelta 
        x_time = tgto.astype('timedelta64[D]').astype(float)#将dimedelta转变为float，这个为因变量

        #因变量初始化处理
        y_low_price,y_ave_price,y_high_price = train_group_tmp_one['最低交易价格'],train_group_tmp_one['平均交易价格'],train_group_tmp_one['最高交易价格']#获取三种因变量值

        #自变量补充并且转化格式
        xnew_time = np.arange(train_group_tmp_one['数据发布时间'][0],train_group_tmp_one['数据发布时间'][len(train_group_tmp_one.index)-1]+pd.Timedelta(days=1),pd.Timedelta(days=1))#插入缺失值的自变量
        #由此化解m8[ns]无法和float运算的问题
        xnew_time_df = pd.DataFrame({'数据发布时间':xnew_time})#np.datetime64变为数据框结构
        xnew_time_df['数据发布时间'] = pd.to_datetime(xnew_time_df['数据发布时间'])#转变为pd.datetime
        xtd = xnew_time_df['数据发布时间'] - xnew_time_df['数据发布时间'][len(xnew_time_df.index)-1] + (train_group_tmp_one['数据发布时间'][len(train_group_tmp_one.index)-1]- train_group_tmp_one['数据发布时间'][0])#计算出timedelta
        xnew_timeF = xtd.astype('timedelta64[D]').astype(float)#timedelta转变为float

        #最低交易价格插值
        f_low_price = interpolate.interp1d(x_time, y_low_price)#最低价格的函数
        ynew_low_price = f_low_price(xnew_timeF)#利用interold进行最低价格的插值

        #平均交易价格插值
        f_ave_price = interpolate.interp1d(x_time, y_ave_price)#最低价格的函数
        ynew_ave_price = f_ave_price(xnew_timeF)#利用interold进行最低价格的插值

        #最高交易价格插值
        f_high_price = interpolate.interp1d(x_time, y_high_price)#最低价格的函数
        ynew_high_price = f_high_price(xnew_timeF)#利用interold进行最低价格的插值

        train_data_final = xnew_time_df
        train_data_final.insert(len(xnew_time_df.columns),value=ynew_low_price,column = '最低交易价格')
        train_data_final.insert(len(xnew_time_df.columns),value=ynew_ave_price,column = '平均交易价格')
        train_data_final.insert(len(xnew_time_df.columns),value=ynew_high_price,column = '最高交易价格')
        return train_data_final#返回一个已经插值的需要训练的一种数据
    
    train_data_finalV = Spline_Interpolation(agrprdValue,markValue)
    
    #将最低和最高价格调整成平均价格，以符合实际
    for i in range(0,len(train_data_finalV['最低交易价格'])):
            if((train_data_finalV['最低交易价格'][i]<1e-15)):
                train_data_finalV.loc[i,('最低交易价格')] = train_data_finalV['平均交易价格'][i]
    for i in range(0,len(train_data_finalV['最高交易价格'])):
            if((train_data_finalV['最高交易价格'][i]<1e-15)):
                train_data_finalV.loc[i,('最高交易价格')] = train_data_finalV['平均交易价格'][i]
      
    #先将预测目标数据935种数据集进行分类
    test_data = data_product_market_dfF1[['农产品市场所在省份','农产品类别','农产品名称映射值','市场名称映射值','数据发布时间']].groupby(by=['农产品市场所在省份','农产品类别','农产品名称映射值','市场名称映射值','数据发布时间'])
    test_group_tmp = test_data.agg('mean')#每个数据只有一个，求和仍是自己

    test_group_tmp.to_csv('tmp/test_group_tmp.csv',encoding='gbk')#保存分类数据为csv
    # test_group_tmp.count(level='农产品名称映射值')
    test_group_tmp2 = pd.read_csv('tmp/test_group_tmp.csv',encoding='gbk')
    test_group_tmp2_df = pd.DataFrame(test_group_tmp2)#在此处选出需要预测的对应的每种菜品
    #获取需要预测的数据
    test_group_tmp_one = pd.DataFrame(columns=test_group_tmp2.columns)
    for i in test_group_tmp2_df.itertuples(index=False):#筛选出一种需要预测的对应菜品
        if (getattr(i, "农产品名称映射值")==agrprdValue)&(getattr(i, "市场名称映射值")==markValue):
            test_group_tmp_one.loc[len(test_group_tmp_one.index)] = i
    test_group_tmp_one['数据发布时间'] = pd.to_datetime(test_group_tmp_one['数据发布时间'])#时间格式转化
    from sklearn.model_selection import train_test_split
    from sklearn.neural_network import MLPRegressor
    #划分训练集和测试集
    data_stand = train_data_finalV['数据发布时间'][0]#以训练数据的最早时间为参考系
    data_tr, data_te, label_tr, label_te = train_test_split(train_data_finalV['数据发布时间'],train_data_finalV[['最低交易价格','平均交易价格','最高交易价格']],test_size=0.2)
    dtr = data_tr - data_stand#计算出timedelta，以训练数据的最早时间为参考系
    data_tr_X = dtr.astype('timedelta64[D]').astype(float)#timedelta转变为float
    dte = data_te - data_stand#计算出timedelta，以训练数据的最早时间为参考系
    data_te_X = dte.astype('timedelta64[D]').astype(float)#timedelta转变为float
    # model = MLPRegressor(
    #     hidden_layer_sizes=(100, ), 
    #     activation='relu', 
    #     solver='adam', 
    #     alpha=0.0001, 
    #     batch_size='auto', 
    #     learning_rate='constant', 
    #     learning_rate_init=0.001, 
    #     power_t=0.5, 
    #     max_iter=1000, 
    #     shuffle=True, 
    #     random_state=None, 
    #     tol=0.0001, 
    #     verbose=False, 
    #     warm_start=False, 
    #     momentum=0.9, 
    #     nesterovs_momentum=True, 
    #     early_stopping=False, 
    #     validation_fraction=0.1, 
    #     beta_1=0.9, 
    #     beta_2=0.999, 
    #     epsilon=1e-08, 
    #     n_iter_no_change=10, 
    #     max_fun=15000)
    model = MLPRegressor(hidden_layer_sizes=(4,2,), activation='relu',solver='lbfgs',max_iter=10000, learning_rate_init=0.2,random_state=100)#构建模型
    model.fit(np.c_[data_tr_X],label_tr)#模型训练
    pre_tr = np.around(np.abs(model.predict(np.c_[data_tr_X])),1)#模型自测
    pre_te = np.around(np.abs(model.predict(np.c_[data_te_X])),1)#模型预测

    err_tr = np.abs(pre_tr - label_tr).mean()
    # err_tr_m = err_tr.mean()
    err_te = np.abs(pre_te - label_te).mean()
#     print(err_tr)
#     print(err_te)
    global err_tr_all
    global err_te_all
    global loss_value
    err_tr_all = np.r_[err_tr_all,err_tr.mean()]
    err_te_all = np.r_[err_tr_all,err_te.mean()]
    loss_value = np.r_[loss_value,model.loss_]
    
#     err_r2 = r2_score(label_tr,pre_tr)
    
    #预测集
    data_fo = test_group_tmp_one['数据发布时间']
    dfo = data_fo - data_stand
    data_fo_X = dfo.astype('timedelta64[D]').astype(float)

    pre_fina = np.around(np.abs(model.predict(np.c_[data_fo_X])),1)
    
    test_group_tmp_one.insert(loc = len(test_group_tmp_one.columns),column='预测最低交易价格',value=pre_fina[:,0])
    test_group_tmp_one.insert(loc = len(test_group_tmp_one.columns),column='预测平均交易价格',value=pre_fina[:,1])
    test_group_tmp_one.insert(loc = len(test_group_tmp_one.columns),column='预测最高交易价格',value=pre_fina[:,2])

    global product_market_pre_tmp
    product_market_pre_tmpsv = pd.concat([product_market_pre_tmp,test_group_tmp_one])#插入数据到总表
    product_market_pre_tmp = pd.DataFrame(product_market_pre_tmpsv)
```
