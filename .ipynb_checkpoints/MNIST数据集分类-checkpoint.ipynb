```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
tf.compat.v1.disable_eager_execution()
#载入数据
mnist = input_data.read_data_sets("MNIST_data",one_hot=True)

#每个批次的大小
batch_size = 1000
#计算一共多少个批次
n_batch = mnist.train.num_examples // batch_size

#定义两个placeholder
x = tf.compat.v1.placeholder(tf.float32,[None,784])
y = tf.compat.v1.placeholder(tf.float32,[None,10])
keep_prob = tf.compat.v1.placeholder(tf.float32)

#创建神经网络
lr = tf.Variable(0.001,dtype=tf.float32)
W1 = tf.Variable(tf.compat.v1.truncated_normal([784,500],stddev = 0.1))
b1 = tf.Variable(tf.zeros([500])+0.1)
L1 = tf.nn.tanh(tf.matmul(x,W1)+b1)
L1_drop = tf.nn.dropout(L1,keep_prob)

W2 = tf.Variable(tf.compat.v1.truncated_normal([500,300],stddev = 0.1))
b2 = tf.Variable(tf.zeros([300])+0.1)
L2 = tf.nn.tanh(tf.matmul(L1_drop,W2)+b2)
L2_drop = tf.nn.dropout(L2,keep_prob)

W3 = tf.Variable(tf.compat.v1.truncated_normal([300,10],stddev = 0.1))
b3 = tf.Variable(tf.zeros([10])+0.1)
prediction = tf.nn.tanh(tf.matmul(L2_drop,W3)+b3)

# W1 = tf.Variable(tf.compat.v1.truncated_normal([784,10]))
# b1 = tf.Variable(tf.zeros([10]))
# prediction = tf.nn.softmax(tf.matmul(x,W)+b)

#二次代价函数
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y,logits=prediction))
#梯度下降法
train_step = tf.compat.v1.train.AdadeltaOptimizer(0.5).minimize(loss)

#初始化变量
init = tf.compat.v1.global_variables_initializer()

#结果存放在一个布尔型列表中
correct_prediction = tf.equal(tf.argmax(y,1),tf.argmax(prediction,1))  #返回一维张量中概率最大的值的位置，并且判断是否相等，相等为true
#求准确率
accuracy = tf.reduce_mean(tf.cast(correct_prediction,tf.float32))  #类型转化，求均值

with tf.compat.v1.Session() as sess:
    sess.run(init)
    for epoch in range(51):
        sess.run(tf.compat.v1.assign(lr,0.001 * (0.95 ** epoch)))
        for batch in range(n_batch):
            batch_xs,batch_ys = mnist.train.next_batch(batch_size)
            sess.run(train_step,feed_dict={x:batch_xs,y:batch_ys,keep_prob:0.0})
        train_acc = sess.run(accuracy,feed_dict={x:mnist.train.images,y:mnist.train.labels,keep_prob:0.0})
        test_acc = sess.run(accuracy,feed_dict={x:mnist.test.images,y:mnist.test.labels,keep_prob:0.0})  #1.0表示神经元全部使用
        
        print("iter"+str(epoch)+", Testing Accuracy "+str(test_acc)+", Traning Accuracy"+str(train_acc)+"Learning Rate"+str(sess.run(lr)))
```

```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
tf.compat.v1.disable_eager_execution()
#载入数据
mnist = input_data.read_data_sets("MNIST_data",one_hot=True)


#每个批次的大小
batch_size = 100
#计算一共多少个批次
n_batch = mnist.train.num_examples // batch_size

def variable_summaries(var):
    with tf.name_scope('summaries'):
        mean = tf.reduce_mean(var)
        tf.compat.v1.summary.scalar('mean',mean)#平均值
        with tf.name_scope('stddev'):
            stddev = tf.sqrt(tf.reduce_mean(tf.square(var-mean)))
        tf.compat.v1.summary.scalar('stddev',stddev)#标准差
        tf.compat.v1.summary.scalar('max',tf.reduce_max(var))#最大值
        tf.compat.v1.summary.scalar('min',tf.reduce_min(var))#最小值
        tf.compat.v1.summary.histogram('histogram',var)#直方图

#命名空间
with tf.name_scope('input'):
    #定义两个placeholder
    x = tf.compat.v1.placeholder(tf.float32,[None,784],name='x-input')
    y = tf.compat.v1.placeholder(tf.float32,[None,10],name='y-input')

with tf.name_scope('layer'):
    #定义两个placeholder
    with tf.name_scope('wights'):
        W = tf.Variable(tf.compat.v1.truncated_normal([784,10]),name='W')
        variable_summaries(W)
    with tf.name_scope('biases'):
        b = tf.Variable(tf.zeros([10]),name='b')
        variable_summaries(b)
    with tf.name_scope('wx_plus_b'):
        wx_plus_b = tf.matmul(x,W)+b
    with tf.name_scope('softmax'):
        prediction = tf.nn.softmax(wx_plus_b)
    

#二次代价函数
with tf.name_scope('loss'):
    loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y,logits=prediction))
    tf.compat.v1.summary.scalar('loss',loss) #报错Fetch的参数不能为空的原因！！！
#梯度下降法
with tf.name_scope('train'):
    train_step = tf.compat.v1.train.GradientDescentOptimizer(0.2).minimize(loss)

#初始化变量
init = tf.compat.v1.global_variables_initializer()

with tf.name_scope('accuracy'):
    #结果存放在一个布尔型列表中
    with tf.name_scope('correct_prediction'):
        correct_prediction = tf.equal(tf.argmax(y,1),tf.argmax(prediction,1))  #返回一维张量中概率最大的值的位置，并且判断是否相等，相等为true
    #求准确率
    with tf.name_scope('accuracy'):
        accuracy = tf.reduce_mean(tf.cast(correct_prediction,tf.float32))  #类型转化，求均值
        tf.compat.v1.summary.scalar('accuracy',accuracy)

#合并所有的summary
merged = tf.compat.v1.summary.merge_all()
        
with tf.compat.v1.Session() as sess:
    sess.run(init)
    writer = tf.compat.v1.summary.FileWriter('logs/',sess.graph)
    for epoch in range(51):
        for batch in range(n_batch):
            batch_xs,batch_ys =  mnist.train.next_batch(batch_size)
            summary,_ = sess.run([merged,train_step],feed_dict={x:batch_xs,y:batch_ys})

        
        writer.add_summary(summary,epoch)
        train_acc = sess.run(accuracy,feed_dict={x:mnist.train.images,y:mnist.train.labels})
        test_acc = sess.run(accuracy,feed_dict={x:mnist.test.images,y:mnist.test.labels})  #1.0表示神经元全部使用
        
        print("iter"+str(epoch)+", Testing Accuracy "+str(test_acc)+", Traning Accuracy"+str(train_acc))
        
#看更多的点的方法(上面不能简单移入第二个for循环，因为全部训练了一次才能生成一个点，而不是训练全部样本中的一个生成一次)
# for i in range(2001):
#     #m每个批次100个样本点
#     batch_xs,batch_ys =  mnist.train.next_batch(100)
#     summary,_ = sess.run([merged,train_step],feed_dict={x:batch_xs,y:batch_ys})
#     writer.add_summary(summary,i)
#     if i%500 == 0:
#         print(sess.run(accuracy,feed_dict={x:mnist.test.images,y:mnist.test.labels}))
```

# Tensorboard可视化

```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
from tensorboard.plugins import projector  #Tensorflow2.0以后的写法
tf.compat.v1.disable_eager_execution()
#载入数据
mnist = input_data.read_data_sets("MNIST_data",one_hot=True)

#运行次数
max_steps = 1001
#图片数量
image_num = 3000
#文件路径
DIR = 'E:/anaconda3_file/projectBN_2/'

#定义会话
sess = tf.compat.v1.Session()

#载入图片
embedding = tf.compat.v1.Variable(tf.compat.v1.stack(mnist.test.images[:image_num]),trainable=False,name='embedding')#横向打包图片

#每个批次的大小
batch_size = 100
#计算一共多少个批次
n_batch = mnist.train.num_examples // batch_size

def variable_summaries(var):
    with tf.compat.v1.name_scope('summaries'):
        mean = tf.compat.v1.reduce_mean(var)
        tf.compat.v1.summary.scalar('mean',mean)#平均值
        with tf.compat.v1.name_scope('stddev'):
            stddev = tf.compat.v1.sqrt(tf.compat.v1.reduce_mean(tf.compat.v1.square(var-mean)))
        tf.compat.v1.summary.scalar('stddev',stddev)#标准差
        tf.compat.v1.summary.scalar('max',tf.compat.v1.reduce_max(var))#最大值
        tf.compat.v1.summary.scalar('min',tf.compat.v1.reduce_min(var))#最小值
        tf.compat.v1.summary.histogram('histogram',var)#直方图

#命名空间
with tf.compat.v1.name_scope('input'):
    #定义两个placeholder
    x = tf.compat.v1.placeholder(tf.compat.v1.float32,[None,784],name='x-input')
    y = tf.compat.v1.placeholder(tf.compat.v1.float32,[None,10],name='y-input')

#显示图片
with tf.compat.v1.name_scope('input_reshape'):
    image_shaped_input = tf.compat.v1.reshape(x,[-1,28,28,1]) #将变量x转变为-1代表未知的，28 28表示将784转化成28*28，维度为1（黑白图片）
    tf.compat.v1.summary.image('input',image_shaped_input,10) #放十张图片
    
with tf.compat.v1.name_scope('layer'):
    #定义两个placeholder
    with tf.compat.v1.name_scope('wights'):
        W = tf.compat.v1.Variable(tf.compat.v1.truncated_normal([784,10]),name='W')
        variable_summaries(W)
    with tf.compat.v1.name_scope('biases'):
        b = tf.compat.v1.Variable(tf.compat.v1.zeros([10]),name='b')
        variable_summaries(b)
    with tf.compat.v1.name_scope('wx_plus_b'):
        wx_plus_b = tf.compat.v1.matmul(x,W)+b
    with tf.compat.v1.name_scope('softmax'):
        prediction = tf.compat.v1.nn.softmax(wx_plus_b)
    

#二次代价函数
with tf.compat.v1.name_scope('loss'):
    loss = tf.compat.v1.reduce_mean(tf.compat.v1.nn.softmax_cross_entropy_with_logits(labels=y,logits=prediction))
    tf.compat.v1.summary.scalar('loss',loss) #报错Fetch的参数不能为空的原因！！！
#梯度下降法
with tf.compat.v1.name_scope('train'):
    train_step = tf.compat.v1.train.GradientDescentOptimizer(0.2).minimize(loss)

#初始化变量
init = tf.compat.v1.global_variables_initializer()

with tf.compat.v1.name_scope('accuracy'):
    #结果存放在一个布尔型列表中
    with tf.compat.v1.name_scope('correct_prediction'):
        correct_prediction = tf.compat.v1.equal(tf.compat.v1.argmax(y,1),tf.compat.v1.argmax(prediction,1))  #返回一维张量中概率最大的值的位置，并且判断是否相等，相等为true
    #求准确率
    with tf.compat.v1.name_scope('accuracy'):
        accuracy = tf.compat.v1.reduce_mean(tf.compat.v1.cast(correct_prediction,tf.float32))  #类型转化，求均值
        tf.compat.v1.summary.scalar('accuracy',accuracy)

#生成metadata文件
if tf.compat.v1.gfile.Exists(DIR + 'projector/projector/metadata.tsv'):
    tf.compat.v1.gfile.DeleteRecursively(DIR + 'projector/projector/metadata.tsv')#存在则删除文件
with open(DIR + 'projector/projector/metadata.tsv','w') as f:
    labels = sess.run(tf.compat.v1.argmax(mnist.test.labels[:],1))#求最大值所在的位置
    for i in range(image_num):
        f.write(str(labels[i])+'\n')
        
#合并所有的summary
merged = tf.compat.v1.summary.merge_all()

projector_writer = tf.compat.v1.summary.FileWriter(DIR + 'projector/projector',sess.graph)#图的结构
saver = tf.compat.v1.train.Saver()#网络模型
config = projector.ProjectorConfig()#定义配置项
embed = config.embeddings.add()
embed.tensor_name = embedding.name
embed.metadata_path = DIR + 'projector/projector/metadata.tsv'
embed.sprite.image_path = DIR + 'projector/data/mnist_10k_sprite.png'
embed.sprite.single_image_dim.extend([28,28])#按照28*28切分
projector.visualize_embeddings(projector_writer,config)#嵌入可视化的工具，放入图结构和配置

for i in range(max_steps):#每次训练100张图片，训练1001次
    #每个批次100个样本
    batch_xs,batch_ys = mnist.train.next_batch(100)
    #两个配置
    run_options = tf.compat.v1.RunOptions(trace_level=tf.compat.v1.RunOptions.FULL_TRACE)
    run_metadatas = tf.compat.v1.RunMetadata()
    summary,_ = sess.run([merged,train_step],feed_dict={x:batch_xs,y:batch_ys},options=run_options,run_metadata=run_metadatas)
    projector_writer.add_run_metadata(run_metadata, 'step%03d' % i)
    projector_writer.add_summary(summary, i)
    
    if i%100 == 0:
        acc = sess.run(accuracy,feed_dict={x:mnist.test.images,y:mnist.test.labels})
        print ("Iter " + str(i) + ", Testing Accuracy= " + str(acc))

saver.save(sess, DIR + 'projector/projector/a_model.ckpt', global_step=max_steps)
projector_writer.close()
sess.close()
```
